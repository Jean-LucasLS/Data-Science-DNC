```python
import pandas as pd

DATA_PATH = 'data/golf_df.xlsx'

df = pd.read_excel(DATA_PATH)
df.head(4)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['temp'] = le.fit_transform(df['temp'])
df['clima'] = le.fit_transform(df['clima'])
df['vento'] = le.fit_transform(df['vento'])
df['umidade'] = le.fit_transform(df['umidade'])
df.head(2)
```

```python
X = df.drop(columns=['jogar'])
y = df['jogar']
```

```python
from sklearn import tree

dtc = tree.DecisionTreeClassifier()
dtc.fit(X, y)
```

```python
fig = tree.plot_tree(dtc, class_names=['yes', 'no'], filled=True)
```
