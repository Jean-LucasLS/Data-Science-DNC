```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
FILE_PATH = 'dados/limite_credito2.xlsx'

df = pd.read_excel(FILE_PATH)
df.head(5)
```

```python
df = pd.get_dummies(df, prefix=['Escolaridade'])
df.head(5)
```

```python
df.corr()
```

```python
df.columns
```

```python
import statsmodels.formula.api as smf

func  = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos'
model = smf.ols(formula=func, data=df).fit()
print(model.summary())
```

```python
import statsmodels.formula.api as smf

func  = 'LimitedoChequeEspecial~Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio - 1'
model = smf.ols(formula=func, data=df).fit()
print(model.summary())
```
