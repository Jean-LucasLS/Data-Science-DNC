```python
import pandas as pd

DATA_PATH = 'data/bank.xlsx'

df = pd.read_excel(DATA_PATH).drop(columns='ID')
df.head(4)
```

```python
df.describe()
```

```python
df['Personal_Loan'].value_counts()
```

```python
X = df.drop(columns=['Personal_Loan'])
y = df['Personal_Loan']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_test  = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(hidden_layer_sizes=(100, 50, 20), activation='logistic')
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score

print(f'\
Accuracy -> {accuracy_score(y_test, clf.predict(X_test)) * 100 :.2f}%\n\
F1_Score -> {f1_score(y_test, clf.predict(X_test)) * 100 :.2f}%')
```
